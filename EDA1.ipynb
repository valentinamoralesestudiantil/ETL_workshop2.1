```python
# Value counts o duplicated data 
# Invoice id or unique identifier for each transaction 
dataframes["spotify_dataset"]["artists"].value_counts()
```

```python
df = dataframes["spotify_dataset"].copy()

artistas_df = df[["artists"]].dropna().copy()

artistas_df["artists_normalizado"] = (
    artistas_df["artists"]
    .str.strip()
    .str.lower()
)

variantes_artistas = (
    artistas_df.groupby("artists_normalizado")["artists"]
    .agg(["nunique", lambda x: list(x.unique())])
    .reset_index()
)

variantes_artistas.columns = ["artists_normalizado", "formas_distintas", "variantes"]
variantes_artistas = variantes_artistas[variantes_artistas["formas_distintas"] > 1]

display(variantes_artistas)
```
